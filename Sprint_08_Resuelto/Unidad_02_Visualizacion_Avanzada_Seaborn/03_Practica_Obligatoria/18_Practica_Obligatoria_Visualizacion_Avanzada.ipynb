### #1.2

Crea una funcin similar a la que se pide en el ejercicio 6 de la segunda tanda de ejercicios del workout. Es decir, una funcin para pintar diagramas de dispersion que tenga seis argumentos, pero ojo al cambio en el quinto:
1. Un dataframe
2. Una columna numrica del dataframe
3. Una columna numrica del dataframe diferente a la anterior.
4. Un columna categrica (Opcional, valor por defecto None)
5. Un valor para el tamao de los puntos o una cadena, pero que puede ser una cadena de caracteres, en cuyo caso se debe interpretar como una de las columnas del dataframe cuyos valores deben controlar el tamao
6. Un sexto argumento "escala", que permita manipular la escala del tamao en el caso de que el quinto argumento sea una cadena. Este argumento multiplicar a los valores de la columna indicada en el parmetro anterior (as si quiero dividir por 10000, este sexto argumento tendr que valer 1/10000)

La funcin debe pintar el diagrama de dispersin de la columna numrica 1, en el eje X, frente a la columna numrica 2, en el eje Y, usando un cdigo de colores en funcin de la columna categrica y con el tamao de los puntos segn el argumento correspondiente. 

Si no se diera una columna categrica como argumento, el color no debe pasarse como parmetro a la funcin de Seaborn o Matplotlib que uses.

EXTRA: Aade un argumento que controle si la funcin pinta una leyenda para colores y tamaos o un colorbar y una leyenda para los tamaos. NO TE PREOCUPES POR EL FORMATO DE LA LEYENDA (lo veremos en la sesin en vivo)

```python
def scatter_custom(
        df,
        col_x,          # columna numrica para eje X
        col_y,          # columna numrica para eje Y
        col_cat=None,   # columna categrica (opcional)
        size=30,        # tamao fijo o nombre de columna
        escala=1,       # escala si size es columna
        mostrar_leyenda=True  # EXTRA
    ):
    # Determinar tamao de los puntos
    if isinstance(size, str):  # size es nombre de columna
        if size not in df.columns:
            raise ValueError(f"La columna '{size}' no existe en el DataFrame.")
        sizes = df[size] * escala
    else:  # size es nmero
        sizes = size

    # Crear figura
    plt.figure(figsize=(10, 6))

    # Si hay columna categrica usamos hue
    if col_cat is not None:
        scatter = sns.scatterplot(
            data=df,
            x=col_x,
            y=col_y,
            hue=col_cat,
            size=None,       # tamaos manejados manualmente abajo
            palette="viridis",
            legend=mostrar_leyenda
        )
    else:
        scatter = sns.scatterplot(
            data=df,
            x=col_x,
            y=col_y,
            color="steelblue",
            size=None,
            legend=False
        )

    # Reemplazar los puntos por sus tamaos reales
    plt.scatter(df[col_x], df[col_y], s=sizes,
                c=df[col_cat] if col_cat is not None else "steelblue",
                cmap="viridis" if col_cat is not None else None)

    # EXTRA
    if col_cat is not None and not mostrar_leyenda:
        plt.colorbar(label=col_cat)

    # Leyenda de tamaos
    if isinstance(size, str) and mostrar_leyenda:
        plt.legend(title=col_cat if col_cat is not None else None)

    # Etiquetas bsicas
    plt.xlabel(col_x)
    plt.ylabel(col_y)
    plt.title("Diagrama de dispersin personalizado")

    plt.show()
```

```python
df = pd.DataFrame(pd.read_csv("./data/titanic.csv"))
df.head()
```

```python
# Diapositiva 1
plt.figure(figsize=(5,4))
sns.countplot(x='survived', data=df)
plt.title("Supervivencia en el Titanic (0=No, 1=S)")
plt.savefig("slide1_supervivencia.png")
plt.show()
```

```python
# Diapositiva 2
plt.figure(figsize=(6,4))
sns.barplot(x='who', y='survived', hue='sex', data=df)
plt.title("Supervivencia por sexo y rol")
plt.savefig("slide2_mujeres_ninos.png")
plt.show()
```

```python
# Diapositiva 3
plt.figure(figsize=(6,4))
sns.barplot(x='class', y='survived', hue='sex', data=df)
plt.title("Supervivencia por clase y sexo")
plt.savefig("slide3_clase.png")
plt.show()
```

```python
# Diapositiva 4
plt.figure(figsize=(6,4))
sns.barplot(x='class', y='survived', hue='who', data=df)
plt.title("Supervivencia por clase y rol")
plt.savefig("slide4_clase_rol.png")
plt.show()
```

```python
# Diapositiva 5
plt.figure(figsize=(6,4))
sns.barplot(x='embark_town', y='survived', hue='class', data=df)
plt.title("Supervivencia segn puerto y clase")
plt.savefig("slide5_puerto.png")
plt.show()
```

```python
# Diapositiva 6
sns.catplot(x='class', y='survived', hue='who', col='embark_town', kind='bar', data=df, height=4, aspect=0.8)
plt.suptitle("Supervivencia combinando clase, rol y puerto", y=1.05)
plt.savefig("slide6_conclusion.png")
plt.show()
```
